```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
df = pd.read_csv("nakamura_1979_sm_locations.csv")
df
```

```python
df = df.drop([2],axis=0)
df = df.drop(["Comments"], axis=1)
df.reset_index(inplace=True, drop=True)
df
```

```python
df['Lat'] = df['Lat'].fillna(-8.9759)
df['Long'] = df['Long'].fillna(45.4986)
df
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Year'].astype(str) + df['Day'].astype(str), format='%Y%j')
df['UTC_datetime'] = df['Date'] + pd.to_timedelta(df['H'], unit='h') + \
                     pd.to_timedelta(df['M'], unit='m') + \
                     pd.to_timedelta(df['S'], unit='s')
df = df.drop(["Year","Day","H","M","S","Date"], axis=1)
df["Type"]="Shallow Moonquake"
df
```

```python
df4 = pd.read_csv("gagnepian_2006_catalog.csv")
df4 = df4.loc[df4['Type']=="M"]
df4.reset_index(inplace=True, drop=True)
df4
```

```python
formatted_dates = [str(date).zfill(10) for date in df4['Date']]
formatted_dates = ['19' + date[:2] + '-' + date[2:4] + '-' + date[4:6] + ' ' + date[6:8] + ':' + date[8:] for date in formatted_dates]
df4['UTCDateTime'] = pd.to_datetime(formatted_dates)
df4
```

```python
df4 = (df4
       .drop(columns = ["YN_Lat","YN_Lon","YN_Depth","Depth_err","Phi","Delta-a","Delta-b","Date"])
       )
df4["Type"] = df4["Type"].str.replace("M","Meteorite Impact")
df4
        
```
